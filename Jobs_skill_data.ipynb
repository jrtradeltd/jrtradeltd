```python
indeed = requests.get(f"https://uk.indeed.com/jobs?q=web%20development&7=")
indeed_soup = bs(indeed.content, 'lxml')
togetto = f"https://uk.indeed.com/viewjob?jk=61d00786607da1ea&tk=1fg9lkirrtv3d802&from=serp&vjs=3"


indeed_jobs = indeed_soup.find("div", id="mosaic-provider-jobcards")
links = indeed_jobs.find_all("a", href=True)
viewjob = f"https://uk.indeed.com/viewjob?"
```

```python
skills = ["Phyton", "Matlab", "Python", "SQL", "Excel", "C++", "AWS", "POWERBI", "QLIK", "VBA", "R", "Django", "Flask", "HTML", "CSS", "JavaScript", "Tkinter" ]
df = pd.DataFrame(columns = ['Job Title'])
for a in indeed_jobs.find_all("a", href=True):
  linktest = a["href"]
  if "/rc" in linktest:
    mylink = viewjob + linktest.strip("/rc/clk?")
    skillrequest = requests.get(f"{mylink}")
    skillrequest_soup = bs(skillrequest.content, 'lxml')
    test = requests.get(f"{mylink}")
    test_soup = bs(test.content, 'lxml')
    jobtitle = test_soup.find("div", class_="jobsearch-JobInfoHeader-title-container")
    jobdesc = test_soup.find("div", id="jobDescriptionText")
    divs = jobdesc.find("div")
    for skill in skills:
      if skill in jobdesc.get_text():
        df = df.append({'Job Title': jobtitle.text, f"{skill}": "yes"}, ignore_index=True)
df
```
